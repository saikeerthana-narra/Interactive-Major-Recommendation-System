```python
def print_centered_text(text):
    # Function to print centered text using HTML formatting
    html = f'<div style="display: flex; justify-content: center; align-items: center; "><h1>{text}</h1></div>'
    
    display(HTML(html))

    
def print_welcome_message(message):
    # Function to print a welcome message with custom formatting
    html = f'<h1 style="font-family: Algerian; font-weight: bold; text-transform: uppercase; text-align: center;font-size: 48px; color: #E84A27;">{message}</h1>'
    
    print_centered_text(html)

    
def print_with_cambria_font(text):
    # Function to print text with Cambria font and center alignment
    html = f'<span style="font-family: Cambria; text-align: center; color: #13294B">{text}</span>'
    
    print_centered_text(html)

def print_with_font(text, font):
    # Function to print text with a specified font
    html = f'<span style="font-family: {font};">{text}</span>'
    
    display(HTML(html))

    
# Call the functions
print_welcome_message("COMPUTING AT ILLINOIS!")

print_with_cambria_font("TAKE THE QUIZ! DISCOVER YOUR PERFECT MAJOR!")


def ask_user_info():
    # Function to ask user for personal information
    name = input("What is your name?")
    
    while True:
        
        try:
            age = int(input("What is your age?"))
            
            break
            
        except ValueError:
            
            print("Invalid input. Please enter a valid number.")
            
            
    print_with_font("Who are you?", "Arial")
    
    print("(1) Student")
    
    print("(2) Parent")
    
    print("(3) Administrator")
    
    print("(4) Other")

    
    while True:
        
        try:
            
            person_choice = int(input("Enter the number corresponding to your choice: "))
            
            if person_choice in [1, 2, 3, 4]:
                
                break
            
            else:
                
                print("Invalid input. Please enter a valid choice.")
        
        except ValueError:
            
            print("Invalid input. Please enter a valid choice.")
       
    
    if person_choice == 1:
        person = "Student"
    
    elif person_choice == 2:
        person = "Parent"
    
    elif person_choice == 2:
        person = "Administrator"
    
    else:
        person = "Other"
    


    print_with_font("What degree are you looking for?", "Arial")
    
    print("(1) Undergraduate")
    
    print("(2) Graduate")
    
    print("(3) Combined BS/MS")
    
    print("(4) PhD/ Doctoral Program")

    
    while True:
        
        try:
            
            area_choice = int(input("Enter the number corresponding to your choice: "))
            
            if area_choice in [1, 2, 3, 4]:
                
                break
            
            else:
                
                print("Invalid input. Please enter a valid choice.")
        
        except ValueError:
            
            print("Invalid input. Please enter a valid choice.")

    
    if area_choice == 1:
        area_of_interest = "Undergraduate Program"
    
    elif area_choice == 2:
        area_of_interest = "Graduate Programs"
    
    elif area_choice == 3:
        area_of_interest = "Combined BS/MS Programs"
    
    else:
        area_of_interest = "PhD/ Doctoral Program"
    

    print_with_font("What is your discipline of interest?", "Arial")
    
    print("(1) Computer Science")
    
    print("(2) Data Science")
    
    print("(3) Computing")
    
    
    while True:
        
        try:
            discipline_choice = int(input("Enter the number corresponding to your choice: "))
            
            if discipline_choice in [1, 2, 3]:
                break
            
            else:
                print("Invalid choice. Please enter a valid option.")   
                
        except ValueError:
            print("Invalid input. Please enter a valid choice.")

    
    if discipline_choice == 1:
        discipline_of_interest = "Computer Science"
    
    elif discipline_choice == 2:
        discipline_of_interest = "Data Science"
    
    else:
        discipline_of_interest = "Computing"
    
    joining = input("When do you want to enroll in the university?")


    
    return name, age, person, area_of_interest, discipline_of_interest, joining
```
